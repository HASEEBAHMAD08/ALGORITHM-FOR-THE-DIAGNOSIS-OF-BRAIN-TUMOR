```python
import numpy as np 
import pandas as pd 
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
breast_cancer=load_breast_cancer()
breast_cancer
```

```python
df=pd.DataFrame(data=breast_cancer.data,columns=breast_cancer.feature_names)
df
```

```python
df["target"]=breast_cancer.target
df['diagnosis']=np.where(df["target"]==0,"Malignant","Benign")
df
```

```python
X=df.drop(["target",'diagnosis'],axis=1)
Y=df["target"]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=20)
```

```python
#Decision tree classifier 
dt_classifier=DecisionTreeClassifier(random_state=35)
dt_classifier.fit(X_train,Y_train)
dt_pred=dt_classifier.predict(X_test)
acc_dt=accuracy_score(Y_test,dt_pred)
acc_dt
```

```python
#Random forest classifier
rf_classifier=RandomForestClassifier(n_estimators=400,random_state=30)
rf_classifier.fit(X_train,Y_train)
rf_pred=rf_classifier.predict(X_test)
acc_rf=accuracy_score(Y_test,rf_pred)
acc_rf
```

```python
def con_matrix(Y_test,Y_pred,title):
    cm=confusion_matrix(Y_test,Y_pred)
    sns.heatmap(cm,annot=True,cmap="Blues",xticklabels=["Malignant","Benign"],yticklabels=["Malignant","Benign"])
    plt.title(title)
```

```python
# confusion matrix for decision tree
con_matrix(Y_test,dt_pred,"con_matrix for decision tree")
```

```python
# confusion matrix for random forest
con_matrix(Y_test,rf_pred,"con_matrix for random forest")
```
